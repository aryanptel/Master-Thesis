```julia
using Plots

"""function logistic_map(r1,r2, x0,y0, n)
    x = zeros(n)
    y = zeros(n)

    x[1] = x0
    y[1] = y0
    sigma = 0.28

    for i in 2:n
        x[i] = r1 * x[i-1] * (1 - x[i-1])
        y[i] = r2* y[i-1] * (1 - y[i-1]) + sigma* (r1 * x[i-1] * (1 - x[i-1]) - r2* y[i-1] * (1 - y[i-1]))
    end

    return x,y
end"""

function logistic_map(r1,r2, x0,y0, n)
    x = zeros(n)
    y = zeros(n)

    x[1] = x0
    y[1] = y0
    epsilon = 0.5

    for i in 2:n
        x[i] = r1 * x[i-1] * (1 - x[i-1]) * (1 - epsilon) + r2 * y[i-1] * (1 - y[i-1]) * (epsilon)
        y[i] = r1 * x[i-1] * (1 - x[i-1]) * (epsilon) + r2 * y[i-1] * (1 - y[i-1]) * (1-epsilon)
    end

    return x,y
end

# Example usage
r1 = 3.9
x0 = 0.1

r2 = r1
y0 = 0.5

num_iterations = 100

x_values, y_values = logistic_map(r1,r2, x0,y0, num_iterations);

# Plotting the logistic map

```

```julia
plot(1:num_iterations, x_values, marker=:circle, linestyle=:solid, label="Logistic Map")
plot!(1:num_iterations, y_values, marker=:circle, linestyle=:solid, label="Logistic Map")

xlabel!("Iteration")
ylabel!("Value")
title!("Logistic Map")
display(Plots.plot!())
```
